### Calculate bidder for highest prebid win ratio 

```python
## We wil create a new DataFrame for our winning ratio to get the highest winning ratio
## As per my understanding to get the  bidder for highest prebid win ratio we can consider reponse count 
## and prebid win count columns
```

```python
df_new=df[['bidder','response_count','prebid_win_count']]
```

```python
df_new.head()
```

```python
df_new.isnull().sum()
```

```python
## Drop Nan Values
```

```python
df_new.dropna(inplace=True)
```

### Group by Bidder and take sum of prebid_win_count and sum of response_count

```python
df_new.head()
```

```python
new_prebid=df_new.groupby('bidder',as_index=False).agg(total_response=('response_count','sum'),total_prebid_win=('prebid_win_count','sum'))
```

```python
new_prebid
```

```python
new_prebid['Winning_percentage']=new_prebid['total_prebid_win']/new_prebid['total_response']
```

```python
new_prebid['Winning_percentage']=new_prebid['Winning_percentage']*100
```

```python
winner_pre_bid=new_prebid.sort_values('Winning_percentage',ascending=False)
```

```python
winner_pre_bid
```

```python
winner_pre_bid.head(1)
```

## Highest winner comes out to be pulsepoint with 79.77% winning ratio 

```python
## Lets check for highest average bidder
```

```python
df_sum_bid=df[['bidder','response_count','sum_bid']]
```

```python
df_sum_bid.isnull().sum()
```

```python
df_sum_bid.dropna(inplace=True)
```

```python
groupd_sum_df=df_sum_bid.groupby('bidder',as_index=False).agg(Total_response_cnt=('response_count','sum'),Total_sum_bid=('sum_bid','sum'))
```

```python
groupd_sum_df['Ratio']=groupd_sum_df['Total_sum_bid']/groupd_sum_df['Total_response_cnt']
```

```python
groupd_sum_df.sort_values('Ratio',ascending=False)
```

```python
## Highest avg bidder also comes out to be pulsepoint with 1.41 avg value/ratio
```
